### Завдання 6 

Напишіть функції, що:
* рохраховує мінімальне та максимальне значення
* середнє значення послідовності
* медіану послідовності
* частоту кожної цифри в послідовності
* середнє квадратичне відхилення

Послідовність чисел ганеруємо рандомно

```python
import random
from math import sqrt

#генеруэмо тисячу чисел в діапазоні від 0 до 10
list_with_random_numbers = [random.randint(0,100) for i in range(1000)]


avg = sum(list_with_random_numbers)/len(list_with_random_numbers)

median_index = (len(list_with_random_numbers)+1)/2

if median_index % 1 == 0.5:
    median = (list_with_random_numbers[int(median_index+0.5)]+list_with_random_numbers[int(median_index-0.5)])/2
else:
    median = list_with_random_numbers[median_index]

standard_deviation_list = [(i-avg)**2 for i in list_with_random_numbers]
std = sqrt(sum(standard_deviation_list)/len(standard_deviation_list))    
    
numbers_dict = {}

for n in list_with_random_numbers:
    numbers_dict[n] = numbers_dict.get(n, 0) + 1

print("Min: ", min(list_with_random_numbers))
print("Max: ", max(list_with_random_numbers))
print("AVG: ", avg) 
print("Median: ", median) 
print("Standard deviation: ", std,"\n")

for k, v in sorted(numbers_dict.items()):
    print (f"Number: {k}, times: {v}")
```
